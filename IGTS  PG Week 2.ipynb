```python
#Q1

"""a) If P2 plays L, P1 best response is U.
      If P2 plays R, P1 best response is D
      If P1 plays U, P2 best response is R
      If P1 plays D, P2 best response is L

   b) p=0.8

   c) q=0.5

   d) { (0.8U,0.2D) , (0.5L,0.5R) } """
```

```python
#Q2

"""
(a)

Let the general 2x2 matrix be represented as:
          Player 2 (Left: q, Right: 1-q)
Player 1    L          R
U (p)     (a, b)     (c, d)
D (1-p)   (e, f)     (g, h)

1. Deriving p* (Player 1's probability) using Player 2's Indifference:
   Player 2 mixes if Expected Payoff(L) = Expected Payoff(R):
   b * p + f * (1 - p) = d * p + h * (1 - p)
   b*p + f - f*p = d*p + h - h*p
   p * (b - f - d + h) = h - f
   p* = (h - f) / (b - d - f + h)

2. Deriving q* (Player 2's probability) using Player 1's Indifference:
   Player 1 mixes if Expected Payoff(U) = Expected Payoff(D):
   a * q + c * (1 - q) = e * q + g * (1 - q)
   a*q + c - c*q = e*q + g - g*q
   q * (a - c - e + g) = g - c
   q* = (g - c) / (a - c - e + g)
"""


def solve_mixed_nash_2x2(matrix):
    a, b = matrix[0][0]  
    c, d = matrix[0][1]  
    e, f = matrix[1][0]  
    g, h = matrix[1][1]  

    p_denom = b - d - f + h
    q_denom = a - c - e + g

    if p_denom == 0 or q_denom == 0:
        print(
            "No strictly interior mixed-strategy Nash Equilibrium exists (Denominator is zero)."
        )
        print("This usually indicates a dominant strategy or multiple pure NE.")
        print("-" * 50 + "\n")
        return None

    p_star = (h - f) / p_denom
    q_star = (g - c) / q_denom

    if not (0 <= p_star <= 1) or not (0 <= q_star <= 1):
        print(
            f"Calculated mathematical bounds fall outside [0, 1]: p*={p_star:.4f}, q*={q_star:.4f}."
        )
        print("No valid interior mixed strategy Nash Equilibrium exists.")
        print("-" * 50 + "\n")
        return None

    print(f"Mixed-Strategy Nash Equilibrium Found:")
    print(f"  -> Player 1 plays Row 0 (U) with probability p* = {p_star:.4f}")
    print(f"  -> Player 2 plays Col 0 (L) with probability q* = {q_star:.4f}")
    print("-" * 50 + "\n")

    return p_star, q_star



#c)
matching_pennies = [[(1, -1), (-1, 1)], [(-1, 1), (1, -1)]]

#ii)
q1_matrix = [[(4, 2), (0, 3)], [(1, 4), (3, 0)]]

#iii)
chicken_game = [[(0, 0), (-1, 1)], [(1, -1), (-10, -10)]]


print("Testing Case (i): Matching Pennies")
solve_mixed_nash_2x2(matching_pennies)

print("Testing Case (ii): Question 1 Matrix")
solve_mixed_nash_2x2(q1_matrix)

print("Testing Case (iii): Chicken Game")
solve_mixed_nash_2x2(chicken_game)
```

```python
#Q3
import random
import matplotlib.pyplot as plt
matching_pennies = [[(1, -1), (-1, 1)], [(-1, 1), (1, -1)]]


def run_simulation(p, q, num_rounds=10000):
    p1_cumulative = 0
    p2_cumulative = 0

    p1_running_averages = []
    p2_running_averages = []

    for r in range(1, num_rounds + 1):
        #a)
        p1_action = 0 if random.random() < p else 1

        p2_action = 0 if random.random() < q else 1

        p1_payoff, p2_payoff = matching_pennies[p1_action][p2_action]

        p1_cumulative += p1_payoff
        p2_cumulative += p2_payoff

        p1_running_averages.append(p1_cumulative / r)
        p2_running_averages.append(p2_cumulative / r)

    return p1_running_averages, p2_running_averages

N = 10000
theoretical_payoff = 0.0 

p1_eq_avg, p2_eq_avg = run_simulation(p=0.5, q=0.5, num_rounds=N)

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(p1_eq_avg, label="Player 1 (Empirical)", color="blue", alpha=0.8)
plt.plot(p2_eq_avg, label="Player 2 (Empirical)", color="orange", alpha=0.8)
plt.axhline(
    y=theoretical_payoff,
    color="red",
    linestyle="--",
    label="Theoretical Expected Payoff (0.0)",
)
plt.title("Part (b): Both Players at Nash Equilibrium\n(p = 0.5, q = 0.5)")
plt.xlabel("Round")
plt.ylabel("Running Average Payoff")
plt.ylim(-1.2, 1.2)
plt.legend()
plt.grid(True, linestyle=":", alpha=0.6)

p1_wrong_avg, p2_wrong_avg = run_simulation(p=0.8, q=0.5, num_rounds=N)

plt.subplot(1, 2, 2)
plt.plot(p1_wrong_avg, label="Player 1 (Empirical)", color="blue", alpha=0.8)
plt.plot(p2_wrong_avg, label="Player 2 (Empirical)", color="orange", alpha=0.8)
plt.axhline(
    y=theoretical_payoff,
    color="red",
    linestyle="--",
    label="Theoretical Expected Payoff (0.0)",
)
plt.title("Part (c): Player 1 Deviates Intentionally\n(p = 0.8, q = 0.5)")
plt.xlabel("Round")
plt.ylabel("Running Average Payoff")
plt.ylim(-1.2, 1.2)
plt.legend()
plt.grid(True, linestyle=":", alpha=0.6)

plt.tight_layout()
plt.show()

"""

(c) ECONOMIC EXPLANATION / COMMENTARY

Even when Player 1 changes their strategy to p = 0.8, the average payoffs for BOTH players still converge to the same theoretical value of 0. This happens because Player 2 is still playing their exact Nash Equilibrium strategy (q* = 0.5), which makes Player 1's expected utility immune to to whatever mixing choices they make. So, Player 1 cannot be penalized nor gain an edge from this as long as Player 2 remains at the 50/50 equilibrium midpoint.
"""
```

```python
#Q4
import matplotlib.pyplot as plt

pd_payoffs = {
    ("C", "C"): [3, 3],
    ("C", "D"): [0, 5],
    ("D", "C"): [5, 0],
    ("D", "D"): [1, 1],
}


#a)
def simulate_backward_induction(T, silent=True):
    """Under strict backward induction in a finitely repeated Prisoner's Dilemma,

    players anticipate mutual defection in the final round T, causing cooperation
    to collapse in round T-1, T-2, and all the way back to round 1.
    Thus, both players play 'D' in every single round.
    """
    total_p1 = 0
    total_p2 = 0

    if not silent:
        print(f"\n--- Detailing Backward Induction Loops (T = {T}) ---")

    for round_num in range(1, T + 1):
        p1_action, p2_action = "D", "D"
        p1_payoff, p2_payoff = pd_payoffs[(p1_action, p2_action)]

        total_p1 += p1_payoff
        total_p2 += p2_payoff

        if not silent:
            print(
                f"  Round {round_num:2d}: P1={p1_action}, P2={p2_action} | Payoffs: P1={p1_payoff}, P2={p2_payoff}"
            )

    return total_p1, total_p2


def simulate_tit_for_tat(T):
    """Tit-for-Tat starts out cooperating in Round 1.

    In every round after that, it mimics whatever the opponent played in the previous round.
    When played against itself, both players cooperate ('C') throughout the entire horizon.
    """
    total_p1 = 0
    total_p2 = 0

    p1_prev = None
    p2_prev = None

    for round_num in range(1, T + 1):
        if round_num == 1:
            p1_action = "C"
            p2_action = "C"
        else:
            p1_action = p2_prev
            p2_action = p1_prev

        p1_payoff, p2_payoff = pd_payoffs[(p1_action, p2_action)]

        total_p1 += p1_payoff
        total_p2 += p2_payoff

        p1_prev = p1_action
        p2_prev = p2_action

    return total_p1, total_p2


#b)
horizons = [5, 20, 100]

bi_results = []
tft_results = []

simulate_backward_induction(T=5, silent=False)

for T in horizons:
    bi_p1, bi_p2 = simulate_backward_induction(T, silent=True)
    tft_p1, tft_p2 = simulate_tit_for_tat(T)

    bi_results.append(bi_p1)
    tft_results.append(tft_p1)

    print(f"\nHorizon T = {T}:")
    print(f"  -> Backward Induction Total Payoff per player: {bi_p1}")
    print(f"  -> Tit-For-Tat Total Payoff per player: {tft_p1}")


#c)
x_indexes = range(len(horizons))
bar_width = 0.35

plt.figure(figsize=(9, 6))

plt.bar(
    [x - bar_width / 2 for x in x_indexes],
    bi_results,
    width=bar_width,
    label="Backward Induction (All Defect)",
    color="crimson",
)
plt.bar(
    [x + bar_width / 2 for x in x_indexes],
    tft_results,
    width=bar_width,
    label="Tit-for-Tat (All Cooperate)",
    color="seagreen",
)

plt.xlabel("Investment Time Horizon (T)")
plt.ylabel("Total Payoff Earned Per Player")
plt.title("Payoff Divergence: Backward Induction vs. Tit-for-Tat")
plt.xticks(x_indexes, [f"T = {t}" for t in horizons])
plt.legend()
plt.grid(axis="y", linestyle=":", alpha=0.6)

plt.tight_layout()
plt.show()

```

```python
#Q5
import numpy as np
import matplotlib.pyplot as plt


#a)
def pv_stream(v, delta, T):
    return sum((delta**t) * v for t in range(T))


def pv_infinite(v, delta):
    return v / (1 - delta)


#b)
def compute_critical_delta(R, T_temptation, P):
    return (T_temptation - R) / (T_temptation - P)


v_test, delta_test, T_test = 6, 0.9, 50
finite_val = pv_stream(v_test, delta_test, T_test)
infinite_val = pv_infinite(v_test, delta_test)

print(f"Finite Horizon PV (T={T_test}): {finite_val:.4f}")
print(f"Infinite Horizon PV:        {infinite_val:.4f}")
print(f"Difference Error Margin:    {abs(infinite_val - finite_val):.4f}\n")

R, T_temptation, P = 3, 5, 1
delta_star = compute_critical_delta(R, T_temptation, P)

print(f"Given parameters: Reward(R)={R}, Temptation(T)={T_temptation}, Punishment(P)={P}")
print(f"Critical Discount Factor (delta*): {delta_star:.4f}\n")

deltas = np.linspace(0.01, 0.99, 500)

pv_cooperate = [pv_infinite(R, d) for d in deltas]

pv_defect = [T_temptation + d * pv_infinite(P, d) for d in deltas]

plt.figure(figsize=(10, 6))
plt.plot(deltas, pv_cooperate, label="PV of Cooperation ($PV_{cooperate}$)", color="seagreen", linewidth=2)
plt.plot(deltas, pv_defect, label="PV of Defection ($PV_{defect}$)", color="crimson", linewidth=2)

plt.axvline(x=delta_star, color="black", linestyle=":", label=f"Critical Threshold ($\delta^*$ = {delta_star:.2f})")
plt.scatter([delta_star], [pv_infinite(R, delta_star)], color="black", zorder=5)

plt.xlabel("Discount Factor ($\delta$) / Patience Level")
plt.ylabel("Present Value of Lifetime Payoffs")
plt.title("Grim Trigger Condition: Cooperation vs. Defection Value")
plt.xlim(0.0, 1.0)
plt.ylim(0, 60)
plt.legend()
plt.grid(True, linestyle=":", alpha=0.6)

plt.tight_layout()
plt.show()
```

```python
#Q6
import matplotlib.pyplot as plt

pricing_payoffs = {
    ("H", "H"): [6, 6],
    ("H", "L"): [0, 9],
    ("L", "H"): [9, 0],
    ("L", "L"): [2, 2],
}

#b)


def strategy_always_defect(history_self, history_opponent):
    """Always chooses the Low price ('L')."""
    return "L"


def strategy_tit_for_tat(history_self, history_opponent):
    """Plays 'H' in round 1, then mimics the opponent's previous move."""
    if not history_opponent:
        return "H"
    return history_opponent[-1]


def strategy_grim_trigger(history_self, history_opponent):
    """Plays 'H' until the opponent plays 'L' once, then plays 'L' forever."""
    if "L" in history_opponent:
        return "L"
    return "H"


#C)


def simulate_game(strategy1_func, strategy2_func, T=100):
    history1, history2 = [], []
    cum_payoffs1, cum_payoffs2 = [0], [0]

    for _ in range(T):
        move1 = strategy1_func(history1, history2)
        move2 = strategy2_func(history2, history1)

        history1.append(move1)
        history2.append(move2)

        p1, p2 = pricing_payoffs[(move1, move2)]
        cum_payoffs1.append(cum_payoffs1[-1] + p1)
        cum_payoffs2.append(cum_payoffs2[-1] + p2)

    return history1, history2, cum_payoffs1[1:], cum_payoffs2[1:]


#d)
T = 100

_, _, h1_cum_ad_tft, h2_cum_ad_tft = simulate_game(
    strategy_always_defect, strategy_tit_for_tat, T
)
print(
    f"Always Defect vs. Tit-for-Tat -> Final Payoffs: AD={h1_cum_ad_tft[-1]}, TFT={h2_cum_ad_tft[-1]}"
)

_, _, h1_cum_ad_gt, h2_cum_ad_gt = simulate_game(
    strategy_always_defect, strategy_grim_trigger, T
)
print(
    f"Always Defect vs. Grim Trigger -> Final Payoffs: AD={h1_cum_ad_gt[-1]}, GT={h2_cum_ad_gt[-1]}"
)

_, _, h1_cum_tft_gt, h2_cum_tft_gt = simulate_game(
    strategy_tit_for_tat, strategy_grim_trigger, T
)
print(
    f"Tit-for-Tat vs. Grim Trigger -> Final Payoffs: TFT={h1_cum_tft_gt[-1]}, GT={h2_cum_tft_gt[-1]}"
)


#e)
rounds = list(range(1, T + 1))
fig, axes = plt.subplots(3, 1, figsize=(8, 12), sharex=True)

axes[0].plot(rounds, h1_cum_ad_tft, label="Always Defect (Firm 1)", color="crimson")
axes[0].plot(rounds, h2_cum_ad_tft, label="Tit-for-Tat (Firm 2)", color="dodgerblue")
axes[0].set_title("Matchup 1: Always Defect vs. Tit-for-Tat")
axes[0].set_ylabel("Cumulative Payoff")
axes[0].legend()
axes[0].grid(True, linestyle=":", alpha=0.6)

axes[1].plot(rounds, h1_cum_ad_gt, label="Always Defect (Firm 1)", color="crimson")
axes[1].plot(rounds, h2_cum_ad_gt, label="Grim Trigger (Firm 2)", color="forestgreen")
axes[1].set_title("Matchup 2: Always Defect vs. Grim Trigger")
axes[1].set_ylabel("Cumulative Payoff")
axes[1].legend()
axes[1].grid(True, linestyle=":", alpha=0.6)

axes[2].plot(rounds, h1_cum_tft_gt, label="Tit-for-Tat (Firm 1)", color="dodgerblue")
axes[2].plot(rounds, h2_cum_tft_gt, label="Grim Trigger (Firm 2)", color="forestgreen")
axes[2].set_title("Matchup 3: Tit-for-Tat vs. Grim Trigger")
axes[2].set_xlabel("Rounds")
axes[2].set_ylabel("Cumulative Payoff")
axes[2].legend()
axes[2].grid(True, linestyle=":", alpha=0.6)

plt.tight_layout()
plt.show()

```
