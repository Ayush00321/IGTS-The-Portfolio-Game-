```python
#Q7
import numpy as np
import matplotlib.pyplot as plt
mean = 0.0008
std = 0.015
dr = np.random.normal(loc=mean, scale=std, size=252)
#print A
ip = 100
ps = ip * np.cumprod(1 + dr)
#part B
plt.figure(figsize=(10, 5))
plt.plot(ps, label="Simulated Stock Price", color="blue", linewidth=2)
plt.title("Simulated Stock Price Series Over 252 Trading Days")
plt.xlabel("Trading Days")
plt.ylabel("Price ($)")
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend()
plt.show()
#part c
print(ps[-1])
print(np.max(ps))
rm = np.maximum.accumulate(ps)
dd = (ps - rm) / rm
print(np.min(dd))
```

```python
import numpy as np
import matplotlib.pyplot as plt
months = ['Jan','Feb','Mar','Apr','May','Jun',
'Jul','Aug','Sep','Oct','Nov','Dec']
year1  =np.array ([42, 45, 50, 48, 55, 60, 58, 63, 67, 70, 75, 80])
year2 = np.array([52, 49, 58, 61, 66, 71, 68, 74, 79, 83, 88, 95])
colors = ['green' if diff >= 0 else 'red' for diff in year1-year2]
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8), sharex=True)
ax1.plot(months, year1, label='Year 1', color='tab:blue', marker='o', linewidth=2)
ax1.plot(months, year2, label='Year 2', color='tab:orange', marker='s', linewidth=2)
ax1.set_title('Monthly Revenue Comparison (in Lakhs)', fontsize=14, fontweight='bold')
ax1.set_ylabel('Revenue (Lakhs)', fontsize=12)
ax1.grid(True, linestyle='--', alpha=0.5)
ax1.legend(fontsize=11)
ax2.bar(months, year1-year2, color=colors, edgecolor='black', alpha=0.8)
ax2.axhline(0, color='black', linewidth=1, linestyle='-')  # Add a baseline at zero
ax2.set_title('Month-by-Month Revenue Growth (Year 2 - Year 1)', fontsize=12, fontweight='bold')
ax2.set_xlabel('Months', fontsize=12)
ax2.set_ylabel('Difference (Lakhs)', fontsize=12)
ax2.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(42)
returns = np.random.multivariate_normal(
mean=[0.0005, 0.0008, 0.0003],
cov=[[0.0004, 0.0001, 0.00005],
[0.0001, 0.0006, 0.00008],
[0.00005, 0.00008, 0.0003]],
size=200
)
prices = 100 * np.cumprod(1 + returns, axis=0)

dfp=pd.DataFrame(prices,columns=['Asset_A', 'Asset_B', 'Asset_C'])


dfp['Portfolio'] = dfp[['Asset_A', 'Asset_B', 'Asset_C']].mean(axis=1)

plt.figure(figsize=(11, 6))
plt.plot(dfp['Asset_A'], label='Asset A', color='tab:blue', alpha=0.7)
plt.plot(dfp['Asset_B'], label='Asset B', color='tab:orange', alpha=0.7)
plt.plot(dfp['Asset_C'], label='Asset C', color='tab:green', alpha=0.7)

dfr=pd.DataFrame(returns,columns=['Asset_A', 'Asset_B', 'Asset_C'])
corrm=dfr.corr()
print(corrm)



plt.plot(dfp['Portfolio'], label='Equally-Weighted Portfolio', 
         color='black', linewidth=2.5, linestyle='--')

plt.title('Simulated Asset & Portfolio Price Series Over 200 Days', fontsize=14, fontweight='bold')
plt.xlabel('Trading Days', fontsize=12)
plt.ylabel('Price ($)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(fontsize=11, loc='upper left')

plt.tight_layout()
plt.show()
```
